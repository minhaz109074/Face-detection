```
import os
import glob
import xml.etree.ElementTree as ET

classes = ['nomask', 'mask']


def xml_to_txt(txt):

  list_with_all_boxes = []
  with open(txt, "r") as f:


       lines = f.readlines()
       des_file = "datasets/" + txt.split(".")[0] + "_face" + ".txt"
       file = open(des_file,'w')
       for line in lines:

          img_name = line.strip().split("/")[2].split(".")[0] 
          
          xml_file = os.path.join("annotations/Pascal_Voc", img_name) + ".xml"
          tree=ET.parse(xml_file)
          root = tree.getroot()
          for obj in root.iter('object'):
            
            cls = obj.find('name').text
            if cls not in classes:
                continue
            cls_id = classes.index(cls)
            xmlbox = obj.find('bndbox')
            list_with_single_boxes = [int(xmlbox.find('xmin').text)-1, int(xmlbox.find('ymin').text)-1, int(xmlbox.find('xmax').text)-1, int(xmlbox.find('ymax').text)-1, int(cls_id)]
            list_with_all_boxes.append(list_with_single_boxes)


          file.write("images/" + img_name + ".jpg")
          file.write('\n')
          file.write('{}'.format(len(list_with_all_boxes)))
          file.write('\n')
          for box in range(len(list_with_all_boxes)):
            data = '{} {} {} {} {}'.format(list_with_all_boxes[box][0], list_with_all_boxes[box][1], list_with_all_boxes[box][2], list_with_all_boxes[box][3], list_with_all_boxes[box][4])
            file.write(data)
            file.write('\n')
          
          list_with_all_boxes = [] 
            

       file.close()
def main():
  file_list = ["train.txt", "test.txt"]
  for file in file_list:
    xml_to_txt(file)
  print("successfully converted to wider_format data")

main()

```
